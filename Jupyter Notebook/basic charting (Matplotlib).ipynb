```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')
x= [1,5,4]
y=[23,45,89]
x2= [3,7,3]
y2= [5,8,9]
plt.plot(x,y, label='first', linewidth= 3)
plt.plot(x2,y2, label='second')
plt.legend() #legend needs to be called seperately
plt.show()
```

for scatter plot

```python
plt.scatter(x,y, label='first', linewidth= 3)
plt.plot(x2,y2, label='second')
plt.legend()
plt.show()
```

for histogram

```python
plt.hist(x, label='first', linewidth= 3)
plt.plot(x2,y2, label='second')
plt.legend()
plt.show()
```
