```python
ndf = df[df.Attendance==0]
ndf
```

```python
links = []

for i in ndf.index:
    year = i[6:10]
    month = i[3:5]
    date = i[0:2]
    
    if (int(year)==1991):
        pos = '100'
        
    elif (int(year)>1991 and int(year)<2002):
        pos = '200'
        
    elif (int(year)>2001 and int(year)<2011):
        pos = '300'
        
    elif (int(year)>2010 and int(year)<2021):
        
        if (int(year)=='2020' and int(month)=='09'):
            pos = '401'
        else:
            pos = '400'
    
    link = f"https://goldsmr4.gesdisc.eosdis.nasa.gov/opendap/MERRA2/M2T1NXAER.5.12.4/{year}/{month}/MERRA2_{pos}.tavg1_2d_aer_Nx.{year}{month}{date}.nc4.nc4?DUCMASS,time,lat,lon"

    links.append(link)
```

```python
links
```

```python
# open file in write mode
with open(r'urls.txt', 'w') as fp:
    for item in links:
        # write each item on a new line
        fp.write("%s\n" % item)
    print('Done')
    
fp.close
```
